## Color Histograms

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dark_horse = cv2.imread('./DATA/horse.jpg')
show_horse = cv2.cvtColor(dark_horse, cv2.COLOR_BGR2RGB)

rainbow = cv2.imread('./DATA/rainbow.jpg')
show_rainbow =cv2.cvtColor(rainbow, cv2.COLOR_BGR2RGB)

blue_bricks = cv2.imread('./DATA/bricks.jpg')
show_bricks = cv2.cvtColor(blue_bricks, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(dark_horse)
```

```python
plt.imshow(show_horse)
```

```python
plt.imshow(show_rainbow)
```

```python
plt.imshow(show_bricks)
```

```python
colors = ['b','g','r']
```

```python
img = dark_horse
for i,col in enumerate(colors):
    histr = cv2.calcHist(images = [img],channels = [i],mask = None,histSize = [256],ranges = [0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.title('Dark Horse')
plt.show()
```

```python
img = blue_bricks
for i,col in enumerate(colors):
    histr = cv2.calcHist(images = [img],channels = [i],mask = None,histSize = [256],ranges = [0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.title('Blue Bricks')
plt.show()
```

```python
img = rainbow
for i,col in enumerate(colors):
    histr = cv2.calcHist(images = [img],channels = [i],mask = None,histSize = [256],ranges = [0,256])
    plt.plot(histr,color = col)
```

```python
def color_hist(img,mask = None):
    colors = ['b','g','r']
    for i,col in enumerate(colors):
        histr = cv2.calcHist(images = [img],channels = [i],mask = mask,histSize = [256],ranges = [0,256])
        plt.plot(histr,color = col)
    
```

## Masked Histogram

```python
images = [rainbow,dark_horse,blue_bricks]
images = [rainbow]

for img in images:
    mask = np.zeros(img.shape[:2],dtype = np.uint8)
    mask[300:400,100:400] = 255
    color_hist(img,mask)
    
```

```python
images = [rainbow,dark_horse,blue_bricks]
images = [rainbow]
for img in images:
    mask = np.zeros(img.shape[:2],dtype = np.uint8)
    mask[300:400,100:400] = 255
    color_hist(img)
    
```

## Histogram Equalization

```python
gorilla = cv2.imread('./DATA/gorilla.jpg',0)
```

```python
def display(img,cmap=None):
    fig = plt.figure(figsize=(10,8))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap = cmap)
```

## Grayscale

```python
display(gorilla,cmap='gray')
```

```python
gorillaHist = cv2.calcHist([gorilla],[0],None,[256],[0,256])
plt.plot(gorillaHist)
plt.title('Gorilla Color Histogram')
plt.show()
```

```python
eqGorilla = cv2.equalizeHist(gorilla)
```

```python
gorillaHist = cv2.calcHist([eqGorilla],[0],None,[256],[0,256])
plt.plot(gorillaHist)
plt.title('Gorilla Color Histogram')
plt.show()
```

```python
display(eqGorilla,cmap = 'gray')
```
