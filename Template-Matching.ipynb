## Template Matching

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
big = cv2.imread('./DATA/sammy.jpg')
big = cv2.cvtColor(big,cv2.COLOR_BGR2RGB)
small = cv2.imread('./DATA/sammy_face.jpg')
small = cv2.cvtColor(small,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(big);
```

```python
plt.imshow(small);
```

```python
methods = ['cv2.TM_SQDIFF','cv2.TM_SQDIFF_NORMED','cv2.TM_CCORR','cv2.TM_CCORR_NORMED','cv2.TM_CCOEFF','cv2.TM_CCOEFF_NORMED']
```

```python
height,width,channels = small.shape
for m in methods:
    big_copy = big.copy()
    method = eval(m)
    res = cv2.matchTemplate(big,small,method)
    minVal,maxVal,minLoc,maxLoc = cv2.minMaxLoc(res)
    
    if method in [cv2.TM_SQDIFF,cv2.TM_SQDIFF_NORMED]:
        bottom_left = minLoc
    else:
        bottom_left = maxLoc
    
    top_right = bottom_left[0] + width,bottom_left[1] + height
    
    cv2.rectangle(big_copy,top_right,bottom_left,color = (255,0,0),thickness = 5)
    plt.subplot(121)
    plt.imshow(res)
    plt.title('Result of Template Matching')
    
    plt.subplot(122)
    plt.imshow(big_copy)
    plt.title('Detected Point')
    plt.suptitle(m)
    
    
    plt.show()
    print('\n')
    print('\n')
```
