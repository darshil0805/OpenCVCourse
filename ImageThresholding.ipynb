## Thresholding on Sudoku Image

```python
img = cv2.imread("./DATA/crossword.jpg",0)
```

```python
def show_pic(img):
    fig = plt.figure(figsize = (15,15))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap = 'gray')
```

```python
show_pic(img)
```

## Applying binary threshold

```python
ret,th1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
show_pic(th1)
```

## Adaptive Thresholding

### OpenCV Thresholding Docs
https://docs.opencv.org/4.5.3/d7/d4d/tutorial_py_thresholding.html

```python
th2 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,13,10)
show_pic(th2)
```

```python
th3 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY,15,8)
show_pic(th3)
```

```python
blended = cv2.addWeighted(src1=th1,alpha=0.7,src2=th2,beta=0.3,gamma=0)
show_pic(blended)
```
