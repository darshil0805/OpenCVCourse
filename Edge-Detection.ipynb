```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

img = cv2.imread('./DATA/sammy_face.jpg')

plt.imshow(img)
```

```python
x = cv2.Canny(img,threshold1 = 100 , threshold2 = 250)
```

```python
plt.imshow(x)
```

```python
med_val = np.median(img)
```

```python
med_val
```

```python
threshold1 = max(0,0.7*med_val)
threshold2 = min(med_val,1.3*med_val)
```

```python
x = cv2.Canny(img,threshold1 = threshold1 , threshold2 = threshold2)
```

```python
plt.imshow(x)
```

```python
blurred_img = cv2.blur(img,ksize=(5,5))
```

```python
x = cv2.Canny(blurred_img,threshold1 = threshold1 , threshold2 = threshold2)
```

```python
plt.imshow(x)
```
